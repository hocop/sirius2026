```python
x = np.arange(0, 10)
print('x =       ', x)
print('x[1:7] =  ', x[1:7])  #    1  6
print('x[1:7:2] =', x[1:7:2])  #    1  6   2
print('x[::2] =  ', x[::2])  #      2
print('x[::-1] = ', x[::-1])  #  . ?
```
